```python
import seaborn as sns
```

```python
ax=sns.countplot(x='Size' ,data=df)
```

```python
ax=sns.countplot(x='Size' ,data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

By going through above graph it could be concluded that size Medium 'M' is the most purchased clothing by people.

In here data is grouped into quantity and size.The groupby() function of pandas is being used.

```python

df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=True)
```

```python
S_Qty=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=True)

sns.barplot(x='Size',y='Qty', data=S_Qty)
```

From above graph it could be concluded that the most number are clothings of size 'M' are bought during sale.

```python

df['Size'].hist() # plotting histograph
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(8, 4))
plt.hist(column_data, bins=20, edgecolor='Yellow')
plt.xticks(rotation=90)
plt.show()
```

As per the above graphical representation we can surely conclude that among the clothing, T-shirts are the one to be purchased by people the most.

```python

B2B_Check = df['B2B'].value_counts() # Checking for B2B Data by using pie chart 

plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.2f%%')
plt.show()
```

```python

plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.2f%%')
plt.show()
```

The above pie graph concludes that the 99.24% of buyers are retailers and the rest 0.76% are B2B.

```python

x_data = df['Category']  # Preparing data for scatter plot graph
y_data = df['Size'] 

plt.scatter(x_data, y_data)
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
```

```python

plt.figure(figsize=(12, 6))   # Plot count of cities by state
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python

top_10_state = df['ship-state'].value_counts().head(10)  # For the top 10 States 

plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

As per the above visuals we can surely conclude that most of the clothings are purchased by people from Maharastra state.

This takes us to a valuable insight to understand that in terms of customers, Maharastra is leading followed by Karnataka and Telangana respectively & with 99.24% of the buyers being retailers, we also found out that T-shirts are in higher demand than others and most prefereed purchased size among buyer is Medium size.
